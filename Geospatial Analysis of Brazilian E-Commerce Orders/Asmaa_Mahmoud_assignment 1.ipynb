```python
my_dayaframe = pd.read_csv("olist_geolocation_dataset.csv")
my_dayaframe.head()
```

```python
limit = 300
my_df_limit = my_dayaframe.iloc[0:limit, :]
my_df_limit.head()
```

```python
# instantiate a feature group for the incidents in the dataframe
new_map = folium.Map(location=[-23.5456212811526, -46.6392920480016], zoom_start=13)

a = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng in zip(my_df_limit.geolocation_lat, my_df_limit.geolocation_lng ):
    a.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
new_map.add_child(a)
```
